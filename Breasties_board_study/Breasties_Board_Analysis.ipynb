```python
import plotly.graph_objects as go

# Define the data
data = {
    'Time Period': ['Initial Survey', 'Intermediate Survey', 'Final Survey'],
    'Never': [45.0, 48.3, 6.7],
    'Occasionally': [33.3, 66.7, 0],
    'Monthly': [57.1, 28.6, 14.3]
}

# Create traces for each group
traces = []
group_names = ['Never', 'Occasionally', 'Monthly']
group_colors = ['rgba(255, 144, 14, 0.5)', 'rgba(255, 65, 54, 0.5)', 'rgba(34, 133, 195, 0.5)']

for i, group_name in enumerate(group_names):
    traces.append(go.Bar(
        x=data['Time Period'],
        y=data[group_name],
        name=group_name,
        marker_color=group_colors[i]
    ))

# Create the layout
layout = go.Layout(
    title='Frequency of Self-check of Breasts',
    xaxis=dict(title='Time Period'),
    yaxis=dict(title='Percentage'),
    barmode='group',
    bargap=0.2  # Adjust the gap between bars within each group
)

# Create the figure
fig = go.Figure(data=traces, layout=layout)

# Show the grouped bar plot
fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
data = {
    'Regular Monitoring': ['Once a month', 'Once every 6 months to a year', 'Every 5 years'],
    'Initial Study': [16.7, 66.7, 16.7],
    'Intermediate Study': [50, 41.7, 8.3],
    'Final Study': [71.4, 28.6, 0]
}

# Extract the x-axis labels and convert them to numerical values
x_labels = data['Regular Monitoring']
x_values = np.arange(len(x_labels))

# Plot the grouped bars for each study
bar_width = 0.2
plt.bar(x_values - bar_width, data['Initial Study'], width=bar_width, label='Initial Study')
plt.bar(x_values, data['Intermediate Study'], width=bar_width, label='Intermediate Study')
plt.bar(x_values + bar_width, data['Final Study'], width=bar_width, label='Final Study')

# Configure the plot
plt.xlabel('Regular Monitoring')
plt.ylabel('Percentage')
plt.title('Regular Monitoring of Breasts')
plt.xticks(x_values, x_labels)
plt.legend()

# Show the plot
plt.show()
```

```python
import plotly.graph_objects as go

# Define the data
data = {
    'Knowing': ['Before', 'After'],
    'Yes': [33.3, 28.6],
    'No': [66.7, 71.4]
}

# Create the grouped bar chart traces
trace1 = go.Bar(
    x=data['Knowing'],
    y=data['Yes'],
    name='Yes',
    marker=dict(color='#1f77b4'),
    text=data['Yes'],
    textposition='auto',
    hovertemplate='Percentage: %{text}%<extra></extra>'
)

trace2 = go.Bar(
    x=data['Knowing'],
    y=data['No'],
    name='No',
    marker=dict(color='#ff7f0e'),
    text=data['No'],
    textposition='auto',
    hovertemplate='Percentage: %{text}%<extra></extra>'
)

# Configure the layout
layout = go.Layout(
    title='Feeling Anxious about Breast Cancer',
    xaxis=dict(title='Knowing about Breasties Board'),
    yaxis=dict(title='Percentage'),
    barmode='group',
    legend=dict(orientation='h', x=0.1, y=-0.25)
)

# Create the figure and add the traces
fig = go.Figure(data=[trace1, trace2], layout=layout)

# Update figure aesthetics
fig.update_layout(
    plot_bgcolor='white',
    font=dict(family='Arial, sans-serif', size=12),
    margin=dict(l=50, r=50, t=80, b=50)
)

fig.update_traces(
    hoverlabel=dict(bgcolor='white', font_size=12, font_family='Arial, sans-serif'),
    textfont=dict(color='black'),
    marker_line_width=1.5
)

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go

# Define the data
data = {
    'Place': ['Public bathroom', 'Health center', 'Workplace'],
    'No. of people': [13, 11, 8]
}

# Create the bar chart trace
trace = go.Bar(
    x=data['Place'],
    y=data['No. of people'],
    marker=dict(color='rgb(63, 127, 191)'),  # Set the color of the bars
    text=data['No. of people'],
    textposition='auto',
    hovertemplate='No. of people: %{text}<extra></extra>',
    width=0.5  # Adjust the width of the bars
)

# Configure the layout
layout = go.Layout(
    title='Preferred Places for Using the Breasties Board',
    xaxis=dict(title='Place'),
    yaxis=dict(title='No. of people'),
    showlegend=False,
    plot_bgcolor='white',
    font=dict(family='Arial, sans-serif', size=12),
    margin=dict(l=80, r=50, t=80, b=50),
    yaxis_tickformat='h',  # Format the y-axis tick labels as integers
    yaxis_range=[0, max(data['No. of people']) * 1.2]  # Set the y-axis range with some padding
)

# Add data labels to the bars
annotations = []
for place, num_people in zip(data['Place'], data['No. of people']):
    annotations.append(dict(
        x=place,
        y=num_people,
        text=str(num_people),
        showarrow=False,
        font=dict(color='black', size=12),
        xanchor='center',
        yanchor='bottom'
    ))
layout['annotations'] = annotations

# Create the figure and add the trace
fig = go.Figure(data=trace, layout=layout)

# Show the plot
fig.show()
```
